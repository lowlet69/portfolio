### Профили покупателей

Для удобства, добавим еще несколько столбцов:
- Стоимость покупки
- Неделя покупки с обозначением даты
- Месяц с обозначением года
- День недели

```python
df['item_total_price'] = df.quantity * df.item_price
df['week'] = df["date"].dt.to_period("W")
df['month'] = df["date"].dt.to_period("M")
df['weekday'] = df["date"].dt.dayofweek
```

Посчитаем средний чек (сначала **полный** чек на каждую покупку, а потом средний чек на каждого покупателя)

```python
#чек на каждую покупку
temp_var = ( #временная переменная
    df
    .groupby('purchase_id')
    .agg({'item_total_price': 'sum'})
    .rename(columns={'item_total_price': 'purchase_price'})
)
temp_var = temp_var.reset_index()
#добавим чек на каждую покупку в наш датасет
df = df.merge(
    temp_var,
    on='purchase_id',
    how='left'
)
```

Посмотрим, что получилось

```python
df.sort_values(by='purchase_id').head(5)
```

Создадим профили пользователей и посчитаем средний чек на каждого

```python
#создаем профили покупателей
profiles = (
    df
    .groupby('user_id')
    .agg({'date': ['first', 'nunique'], #дата первой покупки и количество уникальных покупок
          'item_total_price': 'sum', #lifetime value 
          'purchase_price': 'mean' #средний чек покупателя
         })
)
#убираем мультииндекс
profiles.columns = profiles.columns.droplevel()
#делаем user_id обратно колонкой
profiles = profiles.reset_index()
#переименовываем колонки
profiles = profiles.rename(columns={'first': 'first_dt',
                                    'nunique': 'purchase_count',
                                    'sum': 'ltv',
                                    'mean': 'mean_check'
                                   })
#округляем
profiles['mean_check'] = round(profiles['mean_check'], 2)
```

Добавим маркер участника программы лояльности

```python
#добавляем указание участника программы лояльности
temp_var = ( #временная переменная
    df
    .groupby(['user_id', 'promo'])
    .agg({'quantity': 'sum'})
)
temp_var = temp_var.reset_index()
#соединяем датасеты
profiles = profiles.merge( 
    temp_var[['user_id', 'promo']], 
    on='user_id',
    how='left'
)
```

```python
profiles.head(5)
```

**Вывод:** У нас получились профили пользователей со средним чеком и LTV, которые пригодятся нам для проверки гипотез.

### Оплата за программу лояльности

Для того, чтобы посчитать, сколько заплатили участники программы лояльности за участие в ней, посмотрим на промежуток дат, в которые происходили покупки.

```python
print("Дата начала:", df.date.min())
print("Дата окончания:", df.date.max())
print("Всего дней:", df.date.max() - df.date.min())
```

У нас на три полных месяца приходится 90 дней. Допустим, что покупатель, участвующий в акции, платит за участие каждые 30 дней.

Добавим количество дней, которое покупатель участвует в акции

```python
#сохраним старую переменную
old_profiles = profiles.copy()
```

```python
profiles = (profiles
            #сначала делаем вспомогательный столбец
            .assign(promo_days_1=lambda x: df.date.max() - x.first_dt,
            #чтобы из него получить количество дней в формате числа
                    promo_days=lambda x: x.promo_days_1.dt.days)
            #и избавляемся от него
            .drop(columns={'promo_days_1'})
           )
```

```python
#прописываем условия
conditions = [
    profiles['promo'].eq(0),
    profiles['promo'].eq(1) & profiles['promo_days'].isin(range(1, 31)),
    profiles['promo'].eq(1) & profiles['promo_days'].isin(range(31, 61)),
    profiles['promo'].eq(1) & profiles['promo_days'].isin(range(61, 91)),
]
#прописываем варианты выбора
choices = [0, 200, 400, 600]
#получаем колонку со значением, сколько заплатил покупатель за участие в акции
profiles['promo_value'] = np.select(conditions, choices, default=0)
#прибавляем к ltv значение выше
profiles['ltv'] = profiles['ltv'] + profiles['promo_value']
```

```python
profiles.tail(5)
```

**Примечание:**
- LTV будет или стоять на месте или расти, его нет смысла по датам показывать в презентации, только в дашборде.
- Единственная визуализация, которую здесь вижу - это абсолютные цифры чеков c разбивкой на участников/неучастников программы лояльности (например) за каждую неделю проведения акции. Можно еще выделить отдельно сколько заработали на оплату за программу лояльности.

## Проверка статистических гипотез
Сформулируйте и проверьте статистические гипотезы.
- Проверьте гипотезу: средний чек участников программы лояльности выше, чем у остальных покупателей.
- Сформулируйте собственную статистическую гипотезу. Дополните её нулевой и альтернативной гипотезами. Проверьте гипотезу с помощью статистического теста.

Вторая гипотеза, которую я бы рассмотрел, звучит так: lifetime value участников программы лояльности выше, чем у остальных покупателей. Так как у нас нет расходов на привлечение и есть только цифры выручки, я бы еще попробовал прибавить ещемесячную оплату из 200 рублей к lifetime value.  Итого имеем две нулевых и две альтернативных гипотезы:
1. Гипотеза о среднем чеке
    - $H_{0}$: Между чеками участников программы лояльности и остальными покупателями **нет** значимой разницы.
    - $H_{1}$: Между чеками участников программы лояльности и остальными покупателями **есть** значимая разница.
2. Гипотеза о lifetime value
    - $H_{0}$: Между lifetime value участников программы лояльности и остальными покупателями **нет** значимой разницы.
    - $H_{1}$: Между lifetime value  участников программы лояльности и остальными покупателями **есть** значимая разница.

### Проверка по неотфильтрованным данным

#### Гипотеза о среднем чеке

Применим непараметрический тест Уилкоксона-Манна-Уитни stats.mannwhitneyu(), чтобы защититься от выбросов в данных которые мы видели на графиках. Тест работает с рангами, а не с самими значениями, и сможет показать статистическую значимость даже если данные распределены ненормально.

Гипотеза о среднем чеке
- $H_{0}$: Между чеками участников программы лояльности и остальными покупателями **нет** значимой разницы.
- $H_{1}$: Между чеками участников программы лояльности и остальными покупателями **есть** значимая разница.

```python
#считаем статистическую значимость различий в среднем чеке покупателя между группами
print("p-value: {0:.3f}".format(stats.mannwhitneyu(sample_check_0, sample_check_1)[1]))
#выводим средние показатели
print("Среднее среднего чека без карты лояльности: {:.2f}".format(sample_check_0.mean()))
print("Среднее среднего чека с картой лояльности: {:.2f}".format(sample_check_1.mean()))
#считаем относительную разницу в среднем чеке покупателя между группами
print("Относительная разница: {0:.2%}".format(sample_check_0.mean() / sample_check_1.mean() - 1))
```

**Вывод:** p-value меньше 0.05, поэтому нулевую гипотезу отвергаем. Группы отличаются на 28%. Статистически значимые различия в среднем чеке между группами есть.

#### Гипотеза о lifetime value

Гипотеза о lifetime value
- $H_{0}$: Между lifetime value участников программы лояльности и остальными покупателями **нет** значимой разницы.
- $H_{1}$: Между lifetime value  участников программы лояльности и остальными покупателями **есть** значимая разница.

```python
#считаем статистическую значимость различий в lifetime value покупателя между группами
print("p-value: {0:.3f}".format(stats.mannwhitneyu(sample_ltv_0, sample_ltv_1)[1]))
#выводим средние показатели
print("Среднее lifetimevalue без карты лояльности: {:.2f}".format(sample_ltv_0.mean()))
print("Среднее lifetimevalue с картой лояльности: {:.2f}".format(sample_ltv_1.mean()))
#считаем относительную разницу в lifetime value покупателя между группами
print("Относительная разница: {0:.2%}".format(sample_ltv_1.mean() / sample_ltv_0.mean() - 1))
```
